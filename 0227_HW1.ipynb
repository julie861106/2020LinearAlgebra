# 箭靶

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

i = np.linspace(0, 10, 100)
(x, y) = (0.2*np.sin(i), 0.2*np.cos(i))
(x1, y1) = (np.sin(i), np.cos(i))
(x2, y2) = (2*np.sin(i), 2*np.cos(i))
(x3, y3) = (3*np.sin(i), 3*np.cos(i))
(x4, y4) = (4*np.sin(i), 4*np.cos(i))
(x5, y5) = (5*np.sin(i), 5*np.cos(i))
(x6, y6) = (6*np.sin(i), 6*np.cos(i))
(x7, y7) = (7*np.sin(i), 7*np.cos(i))

plt.style.use("fivethirtyeight")
plt.figure(figsize=(5, 5))
plt.plot(x, y, color = "yellow")
plt.plot(x1, y1, color = "yellow")
plt.plot(x2, y2, color = "red")
plt.plot(x3, y3, color = "red")
plt.plot(x4, y4, color = "blue")
plt.plot(x5, y5, color = "blue")
plt.plot(x6, y6, color = "black")
plt.plot(x7, y7, color = "black")
```
